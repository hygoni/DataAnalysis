```python
from time import sleep
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

size = 10
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
plt.ion()

for i in range(0, len(df)+1 - size):
    df_tmp = df[i:i+size]
    mpl_finance.candlestick2_ohlc(ax, df_tmp['open'], df_tmp['high'], df_tmp['low'], df_tmp['close'],
                                 width = 0.5, colorup='r', colordown='b')
    plt.grid()
    fig.show()
    fig.canvas.draw()
    ax.clear()
    sleep(0.5)
    
```
